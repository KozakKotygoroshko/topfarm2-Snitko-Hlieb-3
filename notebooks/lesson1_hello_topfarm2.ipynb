```python
# Code generated based on Prompt 2 (Copilot Simulation)
colors = ['blue', 'red', 'green', 'yellow']
labels = ['T1', 'T2', 'T3', 'T4']

plt.figure(figsize=(8, 6))
# Loop to plot each point with a different color and label
for i in range(len(x)):
    plt.scatter(x[i], y[i], c=colors[i], label=labels[i], s=100, zorder=2)
    plt.text(x[i]+50, y[i]+50, labels[i], fontsize=12)

plt.title("Layout turbin – Zadanie 2.1 z kolorami i etykietami")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.grid(True, linestyle='--', alpha=0.7)
plt.axis("equal")
plt.legend()
plt.show()
```
