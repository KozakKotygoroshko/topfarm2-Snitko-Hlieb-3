```python
import numpy as np
import matplotlib.pyplot as plt
import topfarm

x = np.array([0, 500, 1000, 1500])
y = np.array([0, 0, 500, 500])

plt.figure(figsize=(8, 5))
plt.scatter(x, y, c='blue', s=100, label='Turbiny')
for i in range(len(x)):
    plt.text(x[i], y[i]+20, f'T{i+1}')
plt.title("Layout turbin – Zadanie 2.1")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.grid(True)
plt.axis("equal")
plt.legend()
plt.show()
```
